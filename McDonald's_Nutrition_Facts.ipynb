```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("menu.csv")
```

```python
# Counting no. of items in each category

sns.countplot(df['Category'])
plt.show()
```

```python
# Finding outliers between different food components

sns.boxplot(df[['Calories','Sugars','Protein','Carbohydrates','Sodium','Dietary Fiber']])
```

```python
plt.figure(figsize=(11,4))
sns.boxplot(df[['Total Fat (% Daily Value)','Cholesterol (% Daily Value)','Carbohydrates (% Daily Value)','Dietary Fiber (% Daily Value)']])
```

```python
# Correlation between data points

corr = df.corr(numeric_only=True)
corr
```

```python
# Heatmap to display the positive and negative correlation

plt.figure(figsize=(15, 10))
sns.heatmap(corr,annot = True)
```

```python
nutrition = pd.pivot_table(df,'Cholesterol (% Daily Value)', index = ['Category'])
nutrition_result = nutrition.sort_values(("Cholesterol (% Daily Value)"),ascending = False)
nutrition_result
```

```python
df['Cholesterol'].idxmax()
```

```python
df.at[31,'Item']
```

```python
#Displaying food categories having higher protein content

highest_protein = df.groupby(['Category'],as_index=False)['Protein'].mean()
sorted = highest_protein.sort_values(by = ['Protein'],ascending = [False])
plt.figure(figsize=(14,5))
plt.title("Categories with higher Protein content")
sns.barplot(x= 'Category' , y = 'Protein' , data = sorted)
plt.show()
```

```python
#Displaying food categories having high calorie content

nutrition = pd.pivot_table(df,'Calories', index = ['Category'])
nutrition_result = nutrition.sort_values(("Calories"),ascending = False)
nutrition_result
```

```python
df['Calories'].idxmax()
```

```python
df.at[82,'Item']
```

```python
highest_fiber = df.groupby(['Category'],as_index=False)['Dietary Fiber'].mean()
sorted = highest_fiber.sort_values(by = ['Dietary Fiber'],ascending = [False])
plt.figure(figsize=(14,5))
plt.title("Categories with higher Fiber content")
sns.barplot(x= 'Category' , y = 'Dietary Fiber' , data = sorted)
plt.show()
```
